**Modelling**

```python
x=df.iloc[:,0:4].values
y=df.iloc[:,4]
```

```python
#Finding best value for 'k'
```

```python
k= range(1,20)
total_acc=[]
for i in k:
  knn=kn(n_neighbors=i,metric="euclidean")
  knn.fit(x,y)
  accr=knn.score(x,y)
  total_acc.append(accr)

plt.figure(figsize=(10,5))
plt.plot(k,total_acc,marker='.')
plt.xlabel("Value of k")
plt.ylabel("Accuracy")
plt.xticks(np.arange(0,20,1))
plt.yticks(np.arange(0.90,1.1,.05))
plt.title("accurate value of k")
plt.show()
```

```python
k=15  #Entering best value of k
pred=[5.1,3.5,41,3.5] #Entering y in order [Sepal length, Sepal width, Petal lenth, Petal width] to predict the class
```

```python
#Training model
```

```python
model=kn(n_neighbors=k,metric="euclidean")
model.fit(x,y)
y_pred=model.predict([pred])
```
